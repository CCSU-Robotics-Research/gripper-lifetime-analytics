```python
import pandas as pd
import matplotlib.pyplot as plt
results = pd.read_csv('airfoil_inspection_5-14-24.csv')

results.head()
```

```python

results['x_diff'] = (results['gripper_x'] - results['part_x']).abs()
results['y_diff'] = (results['gripper_y'] - results['part_y']).abs()
results['r_diff'] = (results['gripper_r'] - results['part_r']).abs()

results.head()
```

```python
cam1_results = results[results['camera_view'] == 1]
cam2_results = results[results['camera_view'] == 2]
cam1_results.head()
```

```python
plt.plot(results.index[results['camera_view'] == 1], results['x_diff'][results['camera_view'] == 1])
plt.xlabel('Index')
plt.ylabel('Relative x Position')
plt.title('Camera View 1: Relative x Position per Cycle')
```

```python
plt.plot(results.index[results['camera_view'] == 1], results['y_diff'][results['camera_view'] == 1])
plt.xlabel('Index')
plt.ylabel('Relative y Position')
plt.title('Camera View 1: Relative y Position per Cycle')
```

```python
plt.plot(results.index[results['camera_view'] == 1], results['r_diff'][results['camera_view'] == 1])
plt.xlabel('Index')
plt.ylabel('Relative Rotation')
plt.title('Camera View 1: Relative Rotation per Cycle')
```

```python
plt.plot(results.index[results['camera_view'] == 2], results['x_diff'][results['camera_view'] == 2], color='red')
plt.xlabel('Index')
plt.ylabel('Relative x Position')
plt.title('Camera View 2: Relative x Position per Cycle')
```

```python
plt.plot(results.index[results['camera_view'] == 2], results['y_diff'][results['camera_view'] == 2], color='red')
plt.xlabel('Index')
plt.ylabel('Relative y Position')
plt.title('Camera View 2: Relative y Position per Cycle')
```

```python
plt.plot(results.index[results['camera_view'] == 2], results['r_diff'][results['camera_view'] == 2], color='red')
plt.xlabel('Index')
plt.ylabel('Relative Rotation')
plt.title('Camera View 2: Relative Rotation per Cycle')
```
