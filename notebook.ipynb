```python
import pandas as pd
import matplotlib.pyplot as plt
results = pd.read_csv('calibrated-data.csv')

results.head()
```

```python

results['x_diff'] = (results['gripper_x'] - results['part_x']).abs()
results['y_diff'] = (results['gripper_y'] - results['part_y']).abs()
results['r_diff'] = (results['gripper_r'] - results['part_r']).abs()

results.head()
```

```python
from sklearn.preprocessing import MinMaxScaler



# Specify which columns to scale
columns_to_scale = ['feature1', 'feature2']

# Create MinMaxScaler object
scaler = MinMaxScaler()

# Fit and transform the specified columns
scaled = scaler.fit_transform(results[columns_to_scale])
plt.plot(results.index[results['camera_view'] == 1], results['gripper_x'][results['camera_view'] == 1], color='blue')
plt.plot(results.index[results['camera_view'] == 1], results['part_x'][results['camera_view'] == 1], color='red')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('r_diff')
plt.title('Comparison of r_diff between Camera Views')

# Add legend
plt.legend()

# Show plot
plt.show()
```
